```
import random 
from matplotlib import pyplot as plt
  

def Rand(n): 
    res = [] 
    for j in range(n): 
      this_tuple = (random.uniform(0, 1), random.uniform(0, 1)) # Generates random x and y coordinates
      res.append(this_tuple)
  
    return res 

# Throw x and y coords into a tuple
  
def main():

  

  list = Rand(10**4)
  # T1=tuple(list)
  print(list)
  # print(T1)

  print("\n\n")
  plt.title('Final orientation')
  

  print("\n\n")
  x,y = zip(*list)
  plt.scatter(x, y)
  # plt.ylabel('Objective Function Value')
  # plt.xlabel('Number of Epochs')
  plt.show()


  

main()




```
